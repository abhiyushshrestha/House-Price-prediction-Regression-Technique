```python
def kgl_submit_csv_keras(df, predictions, savename = "House_Price_prediction",save = "False"):

    print("Prediction", len(predictions))
    
    kgl_df = pd.DataFrame({
        "Id" : df['Id'],
        "SalePrice" : predictions
    })
    
    if save:
        kgl_df.to_csv(savename, index = False)
    return kgl_df
```
